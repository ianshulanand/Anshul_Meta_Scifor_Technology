```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Create a DataFrame from the provided data
data = {
    'Duration': [30, 30, 45, 45, 45, 60, 60, 60, 75, 75],
    'Average_Pulse': [80, 85, 90, 95, 100, 105, 110, 115, 120, 125],
    'Max_Pulse': [120, 120, 130, 130, 140, 140, 145, 145, 150, 150],
    'Calorie_Burnage': [240, 250, 260, 270, 280, 290, 300, 310, 320, 330],
    'Hours_Work': [10, 10, 8, 8, 0, 7, 7, 8, 0, 8],
    'Hours_Sleep': [7, 7, 7, 7, 7, 8, 8, 8, 8, 8]
}

df = pd.DataFrame(data)

# Step 2: Check for missing values
print("Missing values in each column before cleaning:")
print(df.isnull().sum())

# Step 3: Drop rows with missing values (if any)
df_cleaned = df.dropna()

# Step 4: Verify no missing values remain
print("Missing values in each column after cleaning:")
print(df_cleaned.isnull().sum())

# Step 5: Create a bar chart for Average_Pulse
plt.figure(figsize=(8, 6))
sns.barplot(x=df_cleaned.index, y='Average_Pulse', data=df_cleaned)
plt.title('Average Pulse per Entry')
plt.xlabel('Entry Index')
plt.ylabel('Average Pulse')
plt.xticks(rotation=45)
plt.show()

# Step 6: Create a pie chart for Hours_Work
hours_work_counts = df_cleaned['Hours_Work'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(hours_work_counts, labels=hours_work_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Hours Worked')
plt.show()
```

**NumPy**

Use: NumPy is a library for numerical computing in Python. It provides support for arrays (N-dimensional), mathematical functions, and tools for integrating C/C++ and Fortran code.

Example: Creating an array and performing mathematical operations.

import numpy as np

# Create a NumPy array
array = np.array([1, 2, 3, 4, 5])

# Perform mathematical operations
squared_array = array ** 2
print("Squared Array:", squared_array)

# Calculate the mean
mean_value = np.mean(array)
print("Mean Value:", mean_value)

**Matplotlib**

Use: Matplotlib is a plotting library for Python that provides an object-oriented API for embedding plots into applications. It allows for a variety of static, animated, and interactive visualizations.

Example: Creating a line plot.

import matplotlib.pyplot as plt

# Data for plotting
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a line plot
plt.plot(x, y, marker='o')
plt.title('Line Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()

**Seaborn**

Use: Seaborn is built on top of Matplotlib and provides a high-level interface for drawing attractive and informative statistical graphics. It simplifies the creation of complex visualizations and integrates well with Pandas DataFrames.

Example: Creating a bar plot with statistical information.

import seaborn as sns

# Sample data
tips = sns.load_dataset('tips')

# Create a bar plot showing average total bill by day
plt.figure(figsize=(8, 6))
sns.barplot(x='day', y='total_bill', data=tips, estimator=np.mean)
plt.title('Average Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Average Total Bill')
plt.show()
